```python
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import pandas as pd
from bs4 import BeautifulSoup
import chromedriver_autoinstaller
from google.colab import files

chromedriver_autoinstaller.install()

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# disable the images as there are a thousand
prefs = {
    "profile.managed_default_content_settings.images": 2,
    "profile.managed_default_content_settings.stylesheets": 2,
    "profile.managed_default_content_settings.plugins": 2,
    "profile.managed_default_content_settings.popups": 2,
    "profile.managed_default_content_settings.geolocation": 2,
    "profile.managed_default_content_settings.notifications": 2,
    "profile.managed_default_content_settings.media_stream": 2,
}
chrome_options.add_experimental_option("prefs", prefs)


chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3")


driver = webdriver.Chrome(options=chrome_options)
driver.set_page_load_timeout(60)

wait = WebDriverWait(driver, 30)

# login
def login():
    try:
        login_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'a[href="/bruger"]')))
        driver.execute_script("arguments[0].click();", login_button)
        print("Login button clicked.")
    except Exception as e:
        print(f"Error clicking login button: {e}")

    time.sleep(2)

    try:
        email_input = wait.until(EC.visibility_of_element_located((By.NAME, 'email')))
        password_input = driver.find_element(By.NAME, 'password')
        driver.execute_script("arguments[0].value = arguments[1];", email_input, 'grp613@alumni.ku.dk')
        driver.execute_script("arguments[0].value = arguments[1];", password_input, 'xxxx')
        login_submit_button = driver.find_element(By.CSS_SELECTOR, 'a.submit')
        driver.execute_script("arguments[0].click();", login_submit_button)
        print("First login form submitted.")
        time.sleep(5)
    except Exception as e:
        print(f"Error during first login: {e}")

    try:
        premium_login_link = driver.find_element(By.CSS_SELECTOR, 'a[href="/premium/konto?showoverlay"]')
        if premium_login_link:
            premium_login_link.click()
            time.sleep(2)
            premium_email_input = driver.find_element(By.ID, 'premiumLoginEmail')
            premium_password_input = driver.find_element(By.ID, 'premiumLoginPassword')
            driver.execute_script("arguments[0].value = arguments[1];", premium_email_input, 'grp613@alumni.ku.dk')
            driver.execute_script("arguments[0].value = arguments[1];", premium_password_input, 'xxxx')
            premium_login_submit_button = driver.find_element(By.CSS_SELECTOR, 'span.login-button')
            driver.execute_script("arguments[0].click();", premium_login_submit_button)
            print("Premium login form submitted.")
            time.sleep(5)
        else:
            print("No Premium login required.")
    except NoSuchElementException:
        print("Premium login link not found. Proceeding without premium login.")

    try:
        driver.get('https://www.computerworld.dk/bruger')
        try:
            email_input = wait.until(EC.visibility_of_element_located((By.NAME, 'email')))
            password_input = driver.find_element(By.NAME, 'password')
            email_input.send_keys('grp613@alumni.ku.dk')
            password_input.send_keys('xxxx')
            submit_button = driver.find_element(By.CSS_SELECTOR, 'a.submit')
            submit_button.click()
            time.sleep(5)
            print("Logged in successfully via direct URL.")
        except TimeoutException:
            print("Error: Timeout while trying to log in. Please check your login form.")
    except NoSuchElementException as e:
        print(f"Login element not found during direct login: {e}")

# premium paywall login
def handle_premium_paywall_login():
    try:
        premium_login_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'a[href="/premium"]')))
        premium_login_button.click()
        print("Premium login button clicked.")
        time.sleep(2)
        premium_email_input = driver.find_element(By.ID, 'premiumLoginEmail')
        premium_password_input = driver.find_element(By.ID, 'premiumLoginPassword')
        driver.execute_script("arguments[0].value = arguments[1];", premium_email_input, 'grp613@alumni.ku.dk')
        driver.execute_script("arguments[0].value = arguments[1];", premium_password_input, 'xxxx')
        premium_login_submit_button = driver.find_element(By.CSS_SELECTOR, 'span.login-button')
        driver.execute_script("arguments[0].click();", premium_login_submit_button)
        print("Premium credentials submitted.")
        time.sleep(5)
        return True
    except Exception as e:
        print(f"Error during premium paywall login: {e}")
        return False

# cookies and popups
def handle_cookies_and_popups():
    try:
        allow_cookies_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, 'acceptButton')))
        allow_cookies_button.click()
        print("Cookies accepted.")
    except Exception as e:
        print(f"Could not accept cookies: {e}")
    try:
        banner_close_button = driver.find_element(By.CLASS_NAME, 'close')
        driver.execute_script("arguments[0].click();", banner_close_button)
        print("Closed rolloverBanner popup.")
    except Exception as e:
        print("No rolloverBanner popup found or failed to close it.")
    try:
        ad_close_button = driver.find_element(By.CLASS_NAME, 'cw-ad-interstitial-close-button')
        driver.execute_script("arguments[0].click();", ad_close_button)
        print("Closed ad popup.")
    except Exception as e:
        print("No ad popup found or failed to close it.")

# paywall and premium login
def check_paywall_and_relogin():
    try:
        paywall_element = driver.find_element(By.CLASS_NAME, 'premium-article-paywall')
        if paywall_element.is_displayed():
            print("Paywall detected. Re-logging in...")
            if handle_premium_paywall_login():
                driver.refresh()
                time.sleep(5)
                return True
    except NoSuchElementException:
        return False

# scraping articles
def scrape_computerworld_articles(search_term, max_articles=1000):
    article_count = 0
    page = 1
    all_articles = []
    skipped_articles = []

    while article_count < max_articles:
        if page == 0:
            search_url = f"https://www.computerworld.dk/search?q={search_term}"
        else:
            search_url = f"https://www.computerworld.dk/search?q={search_term}#ims;7;0;0;0;{page * 10}/ansigtsgenkendelse"

        driver.get(search_url)
        time.sleep(3 + random.uniform(1, 3))

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        articles = soup.find_all('div', class_='result type article')

        if not articles:
            print("No more articles found.")
            break

        for article in articles:
            if article_count >= max_articles:
                break

            article_url = article.find('a')['href']
            retries = 0
            while retries < 2:
                try:
                    driver.get(article_url)
                    driver.set_page_load_timeout(40)
                    time.sleep(3)

                    if check_paywall_and_relogin():
                        print("Paywall encountered and handled.")
                        driver.refresh()
                        time.sleep(3)

                    article_soup = BeautifulSoup(driver.page_source, 'html.parser')
                    try:
                        title = article_soup.find('h1').get_text()
                    except:
                        title = "No title"
                    try:
                        subheader = article_soup.find('p', class_='teaser').get_text()
                    except:
                        subheader = "No subheader"
                    try:
                        content = " ".join([p.get_text() for p in article_soup.find_all('p', class_='pt')])
                    except:
                        content = "No content"
                    try:
                        date = article_soup.find('span', class_='publishdate').get_text()
                    except:
                        date = "No date"

                    if content.strip() == "":
                        print(f"Warning: Empty content for {article_url}")

                    all_articles.append({
                        'search_term': search_term,
                        'url': article_url,
                        'title': title,
                        'subheader': subheader,
                        'date': date,
                        'content': content.strip()
                    })

                    article_count += 1
                    print(f"Scraped {article_count} articles.")

                    if article_count % 25 == 0:
                        df = pd.DataFrame(all_articles)
                        filename = f'computerworld_scraped_articles_{article_count}.csv'
                        df.to_csv(filename, index=False)
                        print(f"Saved {filename} with {article_count} articles.")
                        files.download(filename)
                    break
                except TimeoutException:
                    print(f"Timeout occurred while loading article: {article_url}. Retrying...")
                    retries += 1
                    if retries == 2:
                        print(f"Skipping article after 2 retries: {article_url}")
                        skipped_articles.append(article_url)
                    continue

        page += 1

    if skipped_articles:
        print("Saving list of skipped articles...")
        df_skipped = pd.DataFrame(skipped_articles, columns=["Skipped_Articles"])
        df_skipped.to_csv('skipped_articles.csv', index=False)
        files.download('skipped_articles.csv')

    return all_articles

driver.get('https://www.computerworld.dk/search?q=ansigtsgenkendelse')
handle_cookies_and_popups()
login()

# search terms to search for on outlets' website
search_terms = ['facial recognition', 'face recognition', 'ansigtsaftryk', 'ansigtsscan',
                'masseovervågning ansigt', 'masseovervågning kamera', 'ansigtskamera', 'overvågningssamfund ansigt',
                'overvågningssamfund kamera', 'sikkerhedspakke overvågning', 'tryghedsskabende kamera']

all_scraped_articles = []

for term in search_terms:
    all_scraped_articles.extend(scrape_computerworld_articles(term))

df = pd.DataFrame(all_scraped_articles)
df.to_csv('computerworld_scraped_articles.csv', index=False)
files.download('computerworld_scraped_articles.csv')

# Close the browser
driver.quit()
```

```python
import re
import pandas as pd

# changing dates
month_mapping = {
    'januar': 'January', 'februar': 'February', 'marts': 'March',
    'april': 'April', 'maj': 'May', 'juni': 'June',
    'juli': 'July', 'august': 'August', 'september': 'September',
    'oktober': 'October', 'november': 'November', 'december': 'December'
}


def clean_date(date_str):
    cleaned_date = re.split(r'kl\. ', date_str)[0].strip()
    for danish_month, english_month in month_mapping.items():
        cleaned_date = cleaned_date.replace(danish_month, english_month)
    return cleaned_date

df['cleaned_date'] = df['date'].apply(clean_date)

df['date'] = pd.to_datetime(df['cleaned_date'], format='%d. %B %Y', dayfirst=True, errors='coerce')

df = df.drop(columns=['cleaned_date'])
df
```
